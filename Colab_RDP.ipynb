```
#@title **RDP Setup**
#@markdown Enter Username and Password

import os
import subprocess
from google.colab import output

username = "woltrex" #@param {type:"string"}
password = "root" #@param {type:"string"}

# Creation of user
os.system(f"useradd -m {username}")
os.system(f"adduser {username} sudo")
os.system(f"echo '{username}:{password}' | sudo chpasswd")
# Using raw string to avoid SyntaxWarning for escape sequences
os.system(r"sed -i 's/\/bin\/sh/\/bin\/bash/g' /etc/passwd")

print("User Created✅.")

#@markdown 1. Open https://remotedesktop.google.com/headless
#@markdown 2. Authenticate and copy the command for **Debian Linux**
#@markdown 3. Paste the command (CRP) below

CRP = "DISPLAY= /opt/google/chrome-remote-desktop/start-host --code=\"4/0AfrIepAE5_s7Pksi8uax7FwrssefXphijpCERwEd7ghIqX92B4an6llLgz9q-hrKQzq9OQ\" --redirect-url=\"https://remotedesktop.google.com/_/oauthredirect\" --name=$(hostname)" #@param {type:"string"}
Pin = 123456 #@param {type: "integer"}

class CRD:
    def __init__(self):
        print("Starting Installation (this takes 4-5 minutes)...")
        self.install_dependencies()
        self.install_crd()
        self.install_desktop()
        self.install_chrome()
        self.finish()

    def install_dependencies(self):
        os.system("apt update")
        os.system("export DEBIAN_FRONTEND=noninteractive")

    def install_crd(self):
        print("Installing Chrome Remote Desktop...")
        subprocess.run(['wget', 'https://dl.google.com/linux/direct/chrome-remote-desktop_current_amd64.deb'], stdout=subprocess.PIPE)
        subprocess.run(['dpkg', '--install', 'chrome-remote-desktop_current_amd64.deb'], stdout=subprocess.PIPE)
        subprocess.run(['apt', 'install', '--assume-yes', '--fix-broken'], stdout=subprocess.PIPE)

    def install_desktop(self):
        print("Installing XFCE4 Desktop...")
        os.system("apt install --assume-yes xfce4 desktop-base xfce4-terminal xscreensaver")
        os.system("bash -c 'echo \"exec /etc/X11/Xsession /usr/bin/xfce4-session\" > /etc/chrome-remote-desktop-session'")
        os.system("systemctl disable lightdm.service")

    def install_chrome(self):
        print("Installing Google Chrome...")
        subprocess.run(["wget", "https://dl.google.com/linux/direct/google-chrome-stable_current_amd64.deb"], stdout=subprocess.PIPE)
        subprocess.run(["dpkg", "--install", "google-chrome-stable_current_amd64.deb"], stdout=subprocess.PIPE)
        subprocess.run(['apt', 'install', '--assume-yes', '--fix-broken'], stdout=subprocess.PIPE)

    def finish(self):
        print("Finalizing configuration...")
        os.system(f"adduser {username} chrome-remote-desktop")
        full_command = f"{CRP} --pin={Pin}"
        os.system(f"su - {username} -c '{full_command}'")
        os.system("service chrome-remote-desktop start")
        print("\nRemote Desktop is Ready! Check https://remotedesktop.google.com/access")

        # Keep-alive logic to help session last longer
        print("\nEnabling Keep-Alive... (Audio loop)")
        display(output.eval_js('new Audio("https://upload.wikimedia.org/wikipedia/commons/0/05/Beep-09.ogg").play()'))

try:
    if not CRP:
        print("❌ Error: Please paste the CRP command from Google Remote Desktop.")
    elif len(str(Pin)) < 6:
        print("❌ Error: Pin must be at least 6 digits.")
    else:
        CRD()
except Exception as e:
    print(f"An error occurred: {e}")
```
